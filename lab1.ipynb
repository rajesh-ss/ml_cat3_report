# 17. simple imputer

```python
df.select_dtypes(include=np.number)
```

```python
df._get_numeric_data()
```

```python
ds.drop("Time", axis=1, inplace=True)
```

```python
ds.drop("Date", axis=1, inplace=True)
```

```python
ds.drop("Invoice ID", axis=1, inplace=True)
```

```python
ds.head()
```

```python
ds['Branch'].replace(['A', 'B', 'C'],[1,2,3],inplace=True)
ds['Gender'].replace(['Female', 'Male'],[0, 1], inplace=True)
ds['City'].replace(['Yangon', 'Naypyitaw', 'Mandalay'],[1, 2, 3], inplace=True)
ds['Customer type'].replace(['Member', 'Normal'],[0, 1], inplace=True)
ds['Payment'].replace(['Ewallet', 'Cash', 'Credit card'],[1, 2, 3], inplace=True)
ds['Product line'].replace(['Health and beauty', 'Electronic accessories', 'Home and lifestyle', 'Sports and travel', 'Food and beverages', 'Fashion accessories'],[1, 2, 3, 4, 5, 6], inplace=True)
```

```python
ds.head()
```

```python
ds['Product line'].unique()
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
# split the data set
numRows = ds.shape[0]
trainSize = round(numRows * 0.8) 
train = ds.iloc[:trainSize, : ]
test = ds.iloc[trainSize : , :]

x_cols = ["City", "Branch","Gender" , "Product line", "Unit price", "Quantity", "Tax 5%", "Total", "Payment", 'cogs', 'gross margin percentage', 'gross income', 'Rating']
y_col = ["Customer type"]

knn.fit(train[x_cols], train[y_col])
```

```python
knn.score(train[x_cols], train[y_col])
```

```python
knn.score(test[x_cols], test[y_col])
```
